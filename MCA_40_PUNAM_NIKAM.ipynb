```python
#Q.1-Write a python program to reverse a string if it is multiple of 5

str1=input("Enter Any string")
if len(str1) % 5 == 0:
    print(str1[::-1])
else:
    print("String is not multiple of 5")
```

```python
#Q.2-Write a python program to identify duplicate characters from the string and reverse it

s = input("Enter any string: ")

dup = ""
for ch in s:
    if s.count(ch) > 1 and ch not in dup:
        dup += ch

print("Duplicates:", dup)
print("Reverse string:", s[::-1])
```

```python
#Q.3-Write a python program to identify digits from string given by the user

str3=input("Enter any string")
if str3.isdigit():
    print("Digits found")
else:
    print("Digits not found")
```
